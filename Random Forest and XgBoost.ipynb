```python
ques_df=new_df[['question1','question2']]
ques_df.head(8)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

questions=list(ques_df['question1'])+list(ques_df['question2'])
cv=CountVectorizer(max_features=3000)
q1_arr,q2_arr=np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
q1_arr
```

```python
q2_arr
```

```python
questions
```

```python
tmp_df1=pd.DataFrame(q1_arr,index=ques_df.index)
tmp_df2=pd.DataFrame(q2_arr,index=ques_df.index)
tmp_df=pd.concat([tmp_df1,tmp_df2],axis=1)
tmp_df.shape
```

```python
tmp_df1
```

```python
tmp_df
```

```python
tmp_df['is_duplicate']=new_df['is_duplicate']
```

```python
tmp_df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(tmp_df.iloc[:,0:-1].values,tmp_df.iloc[:,-1].values,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred=rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
y_pred=xgb.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5) #initating the model with 5 neighbours
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
